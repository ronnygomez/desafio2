```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
data18.head()
```

```python
data18.dtypes
```

```python
#Lista de variables categóricas.
objects = ["Country", "Region"]
```

```python
#Lista de variables numéricas.
numerical = [x for x in data18.columns if x not in objects]
```

```python
#Dataframe con variables numéricas.
data18_numerical = data18[numerical]
data18_numerical.head()
```

```python
#Standarización de las variables.
std_sclr = StandardScaler()
df_std = pd.DataFrame(std_sclr.fit_transform(data18_numerical), index = data18_numerical.index,
                      columns = data18_numerical.columns)
print(df_std.head(5))
```

```python
#Se agrega la variable categórica country al dataframe.
df_std["Country"] = data18["Country"]
```

```python
data18_std = df_std
```

```python
#Se indexa por country.
data18_std.set_index("Country", inplace = True)
```

```python
data18_std.head()
```

```python
#Chequeo de nulos.
data18_std.isnull().sum()
```

```python
data18_std.shape
```

```python
#Se elimina el registro nulo.
data18_std.dropna(axis = 0, inplace = True)
```

```python
#Se instancia el PCA y se hace el dataframe que muestra cuanto aporta cada variable a cada componente.
PCA_data18 = PCA()
PCA_data18.fit(data18_std)
PCA_loadings = pd.DataFrame(PCA_data18.components_.T, index = data18_std.columns, columns = ["PC1", "PC2", "PC3", "PC4",
                                                                                            "PC5","PC6","PC7","PC8"])
PCA_loadings
```

```python
#Se revisa la ortonormalidad entre las componentes.
PCA_loadings.T.dot(PCA_loadings)
```

```python
#Datos originales normalizados en el sistema de coordenadas del espeacio generado por las componentes principales.
df_PCA = pd.DataFrame(PCA_data18.fit_transform(data18_std), columns = range(0, data18_std.shape[1]),
                      index = data18_std.index)
df_PCA.head()
```

```python
#Varianza explicada de cada componente principal.
PCA_data18.explained_variance_
```

```python
#Varianza explicada en porcentaje.
PCA_data18.explained_variance_ratio_
```

```python
#Gráfico de varianza explicada.
plt.figure(figsize = (7, 5))

plt.plot([1, 2, 3, 4, 5, 6, 7, 8], PCA_data18.explained_variance_ratio_, "-o", label = "Componente individual")
plt.plot([1, 2, 3, 4, 5, 6, 7, 8], np.cumsum(PCA_data18.explained_variance_ratio_), "-s", label = "Acumulado")

plt.ylabel("Porcentaje de varianza explicada")
plt.xlabel("Componentes principales")
plt.xlim(0.75, 4.25)
plt.xlim(0, 1.05)
plt.xticks([1, 2, 3, 4, 5, 6, 7, 8])
plt.legend(loc = 2)
```
